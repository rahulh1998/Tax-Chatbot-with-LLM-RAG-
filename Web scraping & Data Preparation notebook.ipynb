```python
import requests
from bs4 import BeautifulSoup

def scrape_and_write_markdown(url, title, target_element="div"):
  """
  Scrapes data from a website and writes it to a markdown file.

  Args:
      url: The URL of the website to scrape.
      title: The title for the markdown file.
      target_element: The HTML element containing the desired data (default: h2).
  """
  try:
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for bad status codes
  except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
    return

  soup = BeautifulSoup(response.content, "html.parser")

  # Find all elements with the target tag
  data_elements = soup.find_all(target_element)

  # print(data_elements)
  # Write markdown content to a file
  with open(f"Website_Data/{title}.md", "w") as f:
    f.write(f"# {title}\n\n")
    for element in data_elements:
      # Extract text and add basic markdown formatting (adjust as needed)
      text = element.text.strip()
      print(text)
      f.write(f"## {text}\n\n")
```

```python
# Scraping the url :  https://y20india.in/income-tax/
scrape_and_write_markdown(url = "https://y20india.in/how-to-efile-income-tax-returns/",
                          title = "How to file income tax returns")
```

```python
## Saving list as a markdown

def save_list_as_markdown(list_name, filename):
  """
  Args:
      list_name: The list to save.
      filename: The output filename.
  """
  with open(f"Data/{filename}.md", "w") as f:
    f.write(f"# {filename}\n\n")

    # Loop through the list items and write them as unordered list items
    for item in list_name:
      f.write(f"* {item}\n")
```

```python
save_list_as_markdown(how_to_file_itr_online,"How To File ITR Online")
save_list_as_markdown(faqs,"FAQ's about filing ITR Online ")
save_list_as_markdown(doc_req,"Documents Required to file an ITR Online ")
```
